# Plot some stocks using yfinance

```python
import yfinance as yf
import quandl as qu
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
data = yf.download(['AAPL', 'SPY'], '2019-01-01', '2020-05-13')
```

```python
data[('Close', 'AAPL')].plot(label='AAPL')
data[('Close', 'SPY')].plot(label='SPY')
plt.title('AAPL and SPY 2019 to 2020-05-13')
plt.legend()
```
